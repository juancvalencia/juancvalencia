```python
# Ajuste de píxeles
plt.rcParams['figure.dpi'] = 100
 # Etiqueta de coordenadas horizontales y verticales
X_label="Precio de Lista"
Y_label="Precio de Venta"

# Datos 1, formato de lista
X_data1 = df['Precio de lista']
Y_data1 = df['Precio de venta']
# Establecer color de capa
_, ax = plt.subplots(facecolor='white', edgecolor='white')

ax.scatter(X_data1, Y_data1, marker='^', s=25, color="r", label="RED")
ax.scatter(X_data1, Y_data1, marker='*', s=12, color="b", label="BLUE")

# Título, etiqueta horizontal y vertical, fuente, tamaño
ax.set_title('Grafico ')
ax.set_xlabel(X_label, fontproperties='Times New Roman', size=12)
ax.set_ylabel(Y_label, fontproperties='Times New Roman', size=12)

plt.legend (loc = 'superior izquierda') # La posición del icono, superior izquierda, superior izquierda, selecciona automáticamente la mejor posición: mejor

plt.show()
```
